# pre-processing of the image

```python
filename = "image_data//b.jpeg"
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(filename, target_size = (224,224))
```

```python
# plt.imshow(img)
```

```python
resized_img = image.img_to_array(img)
final_image =  np.expand_dims(resized_img,axis =0) ## we need 4th dimention of image in DL
fimal_image=tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
final_image.shape
```

```python
prediction = mobile.predict(final_image)
```
